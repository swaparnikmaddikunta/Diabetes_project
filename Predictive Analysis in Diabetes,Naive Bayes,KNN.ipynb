```python
#Performing Data Imputation
df['Pregnancies']=df['Pregnancies'].replace(0,df['Pregnancies'].median())
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].median())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].mean())
df['BMI']=df['BMI'].replace(0,df['BMI'].mean())
df['DiabetesPedigreeFunction']=df['DiabetesPedigreeFunction'].replace(0,df['DiabetesPedigreeFunction'].median())
df['Age']=df['Age'].replace(0,df['Age'].median())
```

3)EDA part-2

```python
#We have to find whether we are having outliers in data or not 
#To do this first let us segregate between input and output parameters
#x-->i/p features and y-->o/p

x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
```

```python
x
```

```python
x.shape
```

```python
y
```

```python
y.shape
```

outlier detection (Generally done using boxplot though we have other methods like empirical formula,z-score etc)

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=x,ax=ax)
plt.savefig('boxplot.jpg')
```

```python
#Detecting outliers
cols=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
for col in cols:
    q1=x[col].quantile(0.25)
    q3=x[col].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    mask=(x[col]<=upper_bound)&(x[col]>=lower_bound)
    
```

```python
#Our focus shouldn't be on complete removal of outliers,we should aim for removal of most of the outliers
#Here we are removing outliers
x_new=x[mask]
y_new=y[mask]
```

```python
x_new.shape#Checking dimension of dataset after removing outliers
```

```python
y_new.shape
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=x_new,ax=ax)
plt.savefig('boxplot.jpg')
```

Concept of Standardization or Normalization.
This is basically done to eliminate bias to particular feature if values are high.Done using 2 methods.They are:-
1)StandardScaler(Here mean=0,standard deviation=1)
2)MinMaxScaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x_new)
```

```python
#Detecting and removing outliers using quantile concept
x_scaled.reset_index(drop=True,inplace=True)
y_new.reset_index(drop=True,inplace=True)

q=x_scaled['Insulin'].quantile(0.95)
mask=x_scaled['Insulin']<q
new_data=x_scaled[mask]
y_new=y_new[mask]
```

5)Model Training

To do training we have to split data into train and test phases
Generally training points are taken more compared to testing one
(8:2 / 7:3 ratio)
We aim for min error rate and high accuracy

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(new_data,y_new,test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.value_counts()
```

Above has data points which are imbalanced--->Data Imbalancing
Mostly observed in health care domain

SMOTE TECHNIQUE

```python
#Applying Smote Technique
#Imb indicates imbalanced class
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train_resampled,y_train_resampled=smote.fit_resample(x_train,y_train)

#Check resampled class distribution
print('Resampled class distribution:')
print(pd.Series(y_train_resampled).value_counts())
```

Model Training:Implementation of Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classification=LogisticRegression()
classification.fit(x_train_resampled,y_train_resampled)
```

Model predictions

```python
y_predict=classification.predict(x_test)
print(y_predict)
```

Model Evaluation Metrics

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

There are some more algorithms like hyperparameter Optimization which can increase this number.
For Health care domain,Recall is the most important parameter.

```python
from sklearn.metrics import classification_report,confusion_matrix
target_names=['Non-Diabetic','Diabetic']
print(classification_report(y_test,y_predict,target_names=target_names))
```

Saving the Model

```python
import pickle
pickle.dump(classification,open('classification_model.pkl','wb'))
```

```python
classification_model=pickle.load(open('classification_model.pkl','rb'))
classification_model.predict(x_test)
```

Naive Bayes (Using Guassian Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
model_guassian_naive_bayes=GaussianNB()
model_guassian_naive_bayes.fit(x_train_resampled,y_train_resampled)
```

```python
y_predict_gaussian_naive_bayes=model_guassian_naive_bayes.predict(x_test)
print(y_predict_gaussian_naive_bayes)
```

```python
#Confusion Matrix
print('Confusion Matrix')
print(confusion_matrix(y_test,y_predict_gaussian_naive_bayes))
```

```python
print('Classification Report')
print(classification_report(y_test,y_predict_gaussian_naive_bayes))
```

```python
print('Accuracy_score')
print(accuracy_score(y_test,y_predict_gaussian_naive_bayes))
```
